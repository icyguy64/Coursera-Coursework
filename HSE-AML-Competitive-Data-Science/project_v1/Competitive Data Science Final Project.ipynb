## Restructure train data to match test data

```python
import numpy as np
from itertools import product
index_cols = ['shop_id', 'item_id', 'date_block_num']

# Iterate through all shops/items combinations
restructured_data = []
for block_num in sales_train['date_block_num'].unique():
    restructured_shops = sales_train.loc[sales_train['date_block_num'] == block_num, 'shop_id'].unique()
    restructured_items = sales_train.loc[sales_train['date_block_num'] == block_num, 'item_id'].unique()
    restructured_data.append(np.array(list(product(*[restructured_shops, restructured_items, [block_num]])),dtype='int32'))
restructured_data = pd.DataFrame(np.vstack(restructured_data), columns = index_cols,dtype=np.int32)
```

```python
restructured_data.head()
```

### There are 50% missing data

```python
test.merge(sales_train, how='left', on=['shop_id', 'item_id']).isnull().sum()
```

### we drop the outliers

```python
sales_train = sales_train[sales_train.item_price<100000]
sales_train = sales_train[sales_train.item_cnt_day<=900]
```

## We have to make predictions by months

```python
sales_train_combined = sales_train.groupby(['date_block_num','shop_id','item_id']).agg({'item_cnt_day': 'sum','item_price': np.mean}).reset_index()
```

## Fill NA entries with 0

```python
sales_train_combined = pd.merge(restructured_data,sales_train_combined,on=['date_block_num','shop_id','item_id'],how='left').fillna(0)
```

## Merge remaining columns item_id and item_category_id

```python
sales_train_combined = pd.merge(sales_train_combined,items,on=['item_id'],how='left')
sales_train_combined = pd.merge(sales_train_combined,item_categories, on=['item_category_id'], how='left')
```

```python
sales_train_combined.head()
```

# Mean encoding

```python
for type_id in ['item_id','shop_id','item_category_id']:
    for column_id, agg_func, func_name in [('item_price',np.mean,'avg'),('item_cnt_day',np.sum,'sum'),('item_cnt_day',np.mean,'avg')]:

        mean_df = sales_train_combined.groupby([type_id,'date_block_num']).agg(agg_func).reset_index()[[column_id,type_id,'date_block_num']]
        mean_df.columns = [type_id+'_'+func_name+'_'+column_id, type_id,'date_block_num']
        
        sales_train_combined = pd.merge(sales_train_combined,mean_df,on=['date_block_num', type_id],how='left')
```

### Attempt to add additional time delayed feature, for this month we are training our data we have previous months data to aid in training

```python
temp_test = test.copy()
temp_test['date_block_num'] = 34
temp_test.drop('ID', axis=1, inplace=True)

temp_test = temp_test.merge(items, how='left', on='item_id')
temp_test = temp_test.merge(item_categories, how='left', on='item_category_id')
temp_test.drop('item_name', axis=1, inplace=True)

sales_train_combined.drop('item_name', axis=1, inplace=True)
sales_train_combined = pd.concat([sales_train_combined,temp_test], axis=0, ignore_index=True)
```

```python
additional_features = ['item_id_avg_item_price',
'item_id_sum_item_cnt_day',
'item_id_avg_item_cnt_day',
'shop_id_avg_item_price',
'shop_id_sum_item_cnt_day',
'shop_id_avg_item_cnt_day',
'item_category_id_avg_item_price',
'item_category_id_sum_item_cnt_day',
'item_category_id_avg_item_cnt_day',
'item_cnt_day']

#features = [1]
features = [1, 2, 3, 4, 5, 6, 9, 12]
for feature in features:
    sales_new_df = sales_train_combined.copy()
    sales_new_df.date_block_num+=feature
    sales_new_df = sales_new_df[['date_block_num','shop_id','item_id']+additional_features]
    sales_new_df.columns = ['date_block_num','shop_id','item_id']+ [delayed_feat+'_delay_'+str(feature) for delayed_feat in additional_features]
    sales_train_combined = sales_train_combined.merge(sales_new_df,on=['date_block_num','shop_id','item_id'] ,how='left')
```

### Fill missing values with zeros

```python
for feat in sales_train_combined.columns:
    if 'item_cnt' in feat:
        sales_train_combined[feat]=sales_train_combined[feat].fillna(0)
    elif 'item_price' in feat:
        sales_train_combined[feat]=sales_train_combined[feat].fillna(sales_train_combined[feat].median())
```

```python
cols_to_drop = additional_features[:-1] + ['item_price'] + ['item_category_name']
sales_train_combined = sales_train_combined[sales_train_combined['date_block_num']>12]
```

# Validation

```python
X_train = sales_train_combined[sales_train_combined['date_block_num']<33].drop(cols_to_drop, axis=1)
X_cv =  sales_train_combined[sales_train_combined['date_block_num']==33].drop(cols_to_drop, axis=1)
X_test = sales_train_combined[sales_train_combined['date_block_num']==34].drop(cols_to_drop, axis=1)
```

# Model Training

```python
params = {
        'eta': 0.08, #best 0.08
        'max_depth': 7,
        'objective': 'reg:linear',
        'eval_metric': 'rmse',
        'seed': 3,
        'gamma':1,
        'silent': True
    }

cols = [c for c in X_train.columns if c not in ['date_block_num', 'item_cnt_day',]]
```

### We make use of XGBoost from scikit learn

```python
import numpy as np
import pandas as pd
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 100)

from itertools import product
from sklearn.preprocessing import LabelEncoder

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from xgboost import XGBRegressor
from xgboost import plot_importance

def plot_features(booster, figsize):    
    fig, ax = plt.subplots(1,1,figsize=figsize)
    return plot_importance(booster=booster, ax=ax)

import time
import sys
import gc
import pickle
```

### Separate the features and predictions

```python
X_training = X_train[cols]
Y_training = X_train['item_cnt_day']
X_valid = X_cv[cols]
Y_valid = X_cv['item_cnt_day']
```

### Start training the model

```python
ts = time.time()

model = XGBRegressor(
    max_depth=8,
    n_estimators=1000,
    min_child_weight=300, 
    colsample_bytree=0.8, 
    subsample=0.8, 
    eta=0.3,    
    seed=42)

model.fit(
    X_training, 
    Y_training, 
    eval_metric="rmse", 
    eval_set=[(X_training, Y_training), (X_valid, Y_valid)], 
    verbose=True, 
    early_stopping_rounds = 10)

time.time() - ts
```

```python
Y_pred = model.predict(X_valid).clip(0, 20)
Y_test = model.predict(X_test[cols]).clip(0, 20)

submission = pd.DataFrame({
    "ID": test.index, 
    "item_cnt_month": Y_test
})
submission.to_csv('submission.csv', index=False)

# save predictions for an ensemble
pickle.dump(Y_pred, open('xgb_train.pickle', 'wb'))
pickle.dump(Y_test, open('xgb_test.pickle', 'wb'))
```
